```python
from PIL import Image
from PIL import ImageFont
from PIL import ImageDraw 
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage.filters import gaussian_filter
```

```python
fonts = [
    ImageFont.truetype("fonts/JetBrainsMono-Medium.ttf", 26),
    ImageFont.truetype("fonts/consolas.ttf", 26),
    ImageFont.truetype("fonts/courier-new.ttf", 26)
]

def createNumpyLetter(letter):
    n =  np.zeros((28,28))
    img = Image.fromarray(n) 
    draw = ImageDraw.Draw(img)
    font = np.random.choice(fonts)
    draw.text(( np.random.randint(-2, 3), np.random.randint(-2, 3)), letter, fill='white', font=font)
    return gaussian_filter(np.array(img.getdata()).reshape((28,28)), sigma=1)
```

```python
plt.figure(figsize=(22, 22))
for i in range(95):
    plt.subplot(8, 12, i + 1)
    plt.imshow(createNumpyLetter(chr(i + 32)), cmap='gray')
```

```python
x_train = [np.array([])]*3000
y_train = [0] * 3000
for i in range(3000):
    n = np.random.randint(0, 96)
    x_train[i] = createNumpyLetter(chr(n + 32))
    y_train[i] = n
#x_train = np.array([createNumpyLetter(chr(i + 32)) for i in range(95)])
#y_train = np.array(range(95))
```

```python
np.savez_compressed('out/x_train', x_train, y_train)
```
