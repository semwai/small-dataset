```python
from PIL import Image
from PIL import ImageFont
from PIL import ImageDraw 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def createNumpyLetter(letter):
    n =  np.zeros((28,28))
    img = Image.fromarray(n) 
    draw = ImageDraw.Draw(img)
    font = ImageFont.truetype("fonts/JetBrainsMono-Medium.ttf", 28)
    draw.text((2,-4), letter, fill='white', font=font)
    return np.array(img.getdata()).reshape((28,28))
```

```python
plt.figure(figsize=(22, 22))
for i in range(95):
    plt.subplot(8, 12, i + 1)
    plt.imshow(createNumpyLetter(chr(i + 32)), cmap='gray')
```

```python
x_train = np.array([createNumpyLetter(chr(i + 32)) for i in range(95)])
y_train = np.array(range(95))
```

```python
np.savez_compressed('out/x_train', x_train, y_train)
```
